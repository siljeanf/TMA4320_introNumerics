## 2.1

Diffusjonslikningen med konstant diffusjonskonstant $D$ tar formen 

\begin{equation}
\label{diffusjonslikningen}
\frac{\partial \phi (x,t) }{\partial t} = D \frac{\partial^2 \phi (x,t) }{\partial x^2}.
\end{equation}

I denne oppgaven skal vi vise at funksjonen $\tilde{\phi}(x,t) = \frac{1} {\sqrt{4 \pi Dt}} e^{-\frac{(x-\mu)^2}{4 D t}}$ er en løsning av diffusjonslikningen \eqref{diffusjonslikningen} over.
 
Setter $\tilde{\phi}(x,t)$ inn i henholdsvis venstre og høyre side av diffusjonslikningen og står igjen med: 

\begin{equation}
\label{venstre}
\frac{\partial \phi (x,t) }{\partial t} = \frac{x^2-2\mu x - 2Dt +\mu^2}{8 \sqrt{\pi t D} D t^2} e^{-\frac{(x-\mu)^2}{4 D t}}
\end{equation}

\begin{equation}
\label{høyre}
D\frac{\partial^2 \phi (x,t) }{\partial x^2} = D \frac{x^2-2\mu x - 2Dt +\mu^2}{8 \sqrt{\pi t D} D^2 t^2}e^{-\frac{(x-\mu)^2}{4 D t}}
\end{equation}



Ser altså at venstre \eqref{venstre} og høyre side \eqref{høyre} er like og vi har dermed vist at $\tilde{\phi}(x,t)$ er en løsning av diffusjonslikningen.

## 2.2.1

Skal her vise at partikkeldistrubisjonen $\phi(x,t)$ ved tiden $t$ er gitt ved følgende integral, 

\begin{equation}
\label{integral_phi}
\phi(x,t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} \mathrm{e}^{-ik(x-x_0)}{e}^{-Dtk^2}\,\mathrm{d}k,
\end{equation}

samt løse  det.


Starter med taylorutviklingen av $\phi(x,t)$ gitt i oppgaveteksten.

\begin{equation}
\phi(x,t) = \phi(x,0) + \left.\sum_{n=1}^\infty \frac{t^n}{n!} \frac{\partial^n \phi (x,t') }{\partial {t'}^n}\right|_{t'=0}.
\label{221_start}
\end{equation}

Finner et uttrykk for $\frac{\partial^n \phi (x,t') }{\partial {t'}^n}$ ved hjelp av diffusjonslikningen,

\begin{equation}
\label{diffusjonslikningen_nte}
\frac{\partial^n \phi (x,t) }{\partial t^n} = D^n \frac{\partial^{2n} \phi (x,t) }{\partial x^{2n}},
\end{equation}

og setter dette inn i \eqref{221_start},

\begin{equation}
\phi(x,t) = \phi(x,0) + \left.\sum_{n=1}^\infty \frac{t^n}{n!} D^n \frac{\partial^{2n} \phi (x,t') }{\partial x^{2n}}\right|_{t'=0}.
\label{221_forts}
\end{equation}

Ser at vi kan få bruk for uttrykket for partikkeldistrubisjonen i t=0 gitt av dirac delta funksjonen:

\begin{equation}
\phi(x,0) = \delta(x-x_0) = \frac{1}{2\pi} \int_{-\infty}^{\infty}dk{e}^{-ik(x-x_0)}.
\label{diracdelta}
\end{equation}

Setter dirac delta funksjonen inn for $\phi(x,t')$ der $t'=0$ i \eqref{221_forts} og forenkler integranden,

\begin{equation}
\phi(x,t) = \phi(x,0) + \sum_{n=1}^\infty \frac{t^n}{n!} D^n \frac{\partial^{2n}}{\partial x^{2n}}\left\{ \frac{1}{2\pi} \int_{-\infty}^{\infty}dk{e}^{-ik(x-x_0)} \right\}
\end{equation}

\begin{equation}
\phi(x,t) = \phi(x,0) + \sum_{n=1}^\infty \frac{t^n}{n!} \frac{D^n}{2\pi} \int_{-\infty}^{\infty} \frac{\partial^{2n}}{\partial x^{2n}} \left\{{e}^{-ik(x-x_0)} \right\}dk
\end{equation}

\begin{equation}
\phi(x,t) = \phi(x,0) + \frac{1}{2\pi} \int_{-\infty}^{\infty} \sum_{n=1}^\infty \frac{(-Dtk^2)^n}{n!}  {e}^{-ik(x-x_0)} dk.
\label{rekke}
\end{equation}

Gjenkjenner den kjente rekken for $e^x$ ved start for $n=0$,

\begin{equation}
e^x = \sum_{n=0}^\infty \frac{x^n}{n!} = 1 + x + \frac{x^2}{2!} +...
\label{e}
\end{equation}

og forenkler \eqref{rekke} ytterligere:

\begin{equation}
\phi(x,t) = \phi(x,0) + \frac{1}{2\pi} \int_{-\infty}^{\infty}({e}^{-Dtk^2}-1) {e}^{-ik(x-x_0)} dk
\end{equation}

\begin{equation}
\phi(x,t) = \phi(x,0) - \frac{1}{2\pi} \int_{-\infty}^{\infty} {e}^{-ik(x-x_0)} dk + \frac{1}{2\pi}\int_{-\infty}^{\infty}({e}^{-ik(x-x_0) - Dtk^2}dk
\label{snartferdig}
\end{equation}

Gjenkjenner og bytter ut første integral i \eqref{snartferdig} med $\phi(x,0)$ gitt av dirac delta funksjonen \eqref{diracdelta}

\begin{equation}
\phi(x,t) = \phi(x,0) - \phi(x,0) + \frac{1}{2\pi}\int_{-\infty}^{\infty}({e}^{-ik(x-x_0) - Dtk^2}dk
\end{equation}

\begin{equation}
\phi(x,t) =  \frac{1}{2\pi}\int_{-\infty}^{\infty}({e}^{-ik(x-x_0) - Dtk^2}dk
\label{løsning_int}
\end{equation}

Har dermed løst uttrykket for partikkeldistrubisjonen $\phi(x,t)$ ved hjelp av initialbetingelser for $t=0$.
Neste steg er å løse integralet over. Legger merke til at vi kan bruke det fullstendige kvadrat \eqref{fullstendigkvadrat} til å rydde opp i eksponenten før den settes inn med resten av integranden i \eqref{løsning_int}:

\begin{equation}
a^2 + 2ab + b^2 = (a+b)^2
\label{fullstendigkvadrat}
\end{equation}

\begin{equation}
-exp = Dtk^2 + ik(x- x_0) = Dtk^2 + (x-x_0)ik + \frac {(x-x_0)^2 (ik)^2 }{(4Dtk)} - \frac {(x-x_0)^2 (ik)^2 }{(4Dtk)}
\end{equation}

\begin{equation}
-exp = \Big( \sqrt{Dtk^2} + \frac {(x-x_0)i}{ \sqrt {(4Dt)} } \Big )^2 - \frac {i(x-x_0)^2  }{(4Dt)}.
\end{equation}

Setter så inn den faktoriserte eksponenten i integranden i \eqref{løsning_int},

\begin{equation}
\phi(x,t) = \frac{1}{2\pi} {e}^{-\frac{(x-x_0)^2}{4Dt}}\int_{-\infty}^{\infty} {e}^{ - \Big(\sqrt{Dt}k + \frac{ (x-x_0)^2 }{4Dt})\Big)}.
\end{equation}
 
Ved hjelp av substitusjon for $u = \sqrt{Dt}k + \frac{(x-x_0)^2}{4Dt}$ får vi 

\begin{equation}
\phi(x,t) = {e}^{-\frac{(x-x_0)^2}{4Dt}} \frac{1}{2\pi} \int_{-\infty}^{\infty} {e}^{ - u^2} \frac{du}{\sqrt{Dt}}.
\end{equation}
 
Tar i bruk gaussintegralet:

\begin{equation}
\int_{-\infty}^{\infty} {e}^{ - x^2}dx = \sqrt{\pi}.
\label{gauss}
\end{equation}

Det gir oss det endelige resultatet av integralet:
 
\begin{equation}
\phi(x,t) = \frac{1}{\sqrt {4 \pi Dt}} e^{-\frac{(x-x_0)^2}{4Dt}}
\label{ferdig}
\end{equation}


## 2.2.2

#### Fysisk tolkning av diffusjonskonstanten 

Legger merke til at løsningen for partikkeldistrubisjonen, $\phi(x,t)$, er normalfordelt for partikkelposisjonen, $x$. En standard normalfordeling med forventningsverdi $\mu$ og standardavvik $\sigma$ tar formen

\begin{equation}
f(x) = \frac{1}{2\pi \sigma^2} {e}^{-\frac{(x-\mu)^2}{(2\sigma^2)}}.
\end{equation}

Dette gir standardavviket $\sigma = \sqrt{2Dt}$ i vårt tilfelle. Vi vet at standardavviket, $\sigma$, er et mål på spredning, altså et mål på hvor langt måleverdiene er unna gjennomsnittsverdien. Vi ønsker å undersøke hvordan partikkeldistrubisjonen $\phi(x,t)$ varierer ved økende D. Uttrykket for $\sigma$
er omvendt proporsjonalt med roten av tiden $t$ og diffusjonskonstanten $D$,

\begin{equation}
\phi \sim \frac{1}{\sqrt{Dt}}.
\label{forhold}
\end{equation}

Det er logisk at $\phi$ blir stadig mer normalfordelt, og at fordelingen blir jevnere med økende tid $t$. Dersom vi øker diffusjonskonstanten, $D$, vil partikkeldistrubisjonen raskere bli normalfordelt ved samme antall tidssteg. Dette gjør at $D$ kan sees på som et mål på spredning i likhet med $t$. Allikevel er det verdt å merke seg at $t$ og $D$ vokser som $\sqrt{t}$ og $\sqrt{D}$, slik at økningen vil gå relativt sakte. Det at standardavviket vokser veldig tregt er et karakteristisk trekk for diffusjonsprosesser.

#### Oppgaven omhandler virrevandring i 1D


I denne oppgaven har vi simulert virrevandring i 1D med $n = 1000$ partikler, som slippes i $x = 0$. Virrevandringen gjennomføres over $100$ tidssteg. En partikkel kan forflytte seg til venstre eller høyre. Vi antar at sannsynligheten for at en partikkel beveger seg til venstre/høyre er $0.5$, og at sannsynligheten for at partikkelen står stille er $0$. 

Virrevandringen er simulert i funksjonen random_walk. Denne funksjonen tar inn antall partikler *n,* sannsynligheten for hva det neste steget til partikkelen blir *prob*, og antall tidssteg simulasjonen skal gå over, *timestep*. Det genereres et tilfeldig tall mellom $0$ og $1$ for hver paritkkel ved hvert tidssteg som bestemmer hvilken vei partikkelen bevever seg gitt sannsynligheten *prob*.
Funksjonen returnerer en vektor, *pos*, som inneholder posisjonen til alle partiklene i systemet ved siste tidssteg, forventningsverdi og standardavvik til posisjonen til en partikkel.

```python
n3 = 1000 #antall partikler i x=0
prob3 = 0.5 #sannsynlighet for at partikkelen  beveger seg til venstre/høyre
h = 1
timestep3 = 100
 

def random_walk(n, prob, timestep): #Finner partikkeldistrubisjon etter timestep tidssteg
    pos = np.zeros(n) #vektor med posisjonene til n partikler
    for i in range(timestep): #antall tidssteg
        for j in range(n): #antall partikler
            rr = np.random.uniform() #genererer et tilfeldig tall mellom 0.0 og 1.0
            if (rr >= prob): #hvis det tilfeldige tallet >= 0.5, forflytter partikkelen seg et steg til høyre
                pos[j] += h
            else: #hvis det tilfeldige tallet < 0.5, forflytter partikkelen seg et steg til venstre
                pos[j] -= h
                
    mean, std = norm.fit(pos) #beregner forventingsverdi og standardavvik til posisjonen til partiklene
    
    return pos, mean, std
```

```python
#gjennomfører simulering av virrevandring med 1000 partikler, sanns. = 0.5 over 100 tidssteg
pos1, mean1, std1 = random_walk(n3,prob3, timestep3)
```

```python
# Plotter histogram over partikkeldistribusjon
plt.figure(figsize = (10,6))
plt.title('Virrevandringer i 1D')
n1, bins1, patches1 = plt.hist(pos1, bins = range(-50,50,1), label = "Partikkeldistribusjon") 
plt.plot(bins1, norm.pdf(bins1, mean1, std1)*n3*2, label = "Normalfordeling")
plt.xlabel("Posisjon")
plt.ylabel("Antall partikler i hver posisjon")
plt.legend()
plt.show()
```

#### Denne oppgaven omhandler virrevandring i potensial

```python
#Ulike konstanter og verdier som brukes i oppgave 5

T = 310 # temperatur målt i Kelvin
kB = 1.38064852e-23 # Boltzmanns konstant målt i J/K
beta = 1/(kB*T)

k1 = 0.1/beta 
k2 = 1/beta
k3 = 10/beta

h = 1 #steglengde
n5 = 1000 #antall partikler
timestep5 = 100 #antall tidssteg
```

I denne oppgaven har vi simulert virrevandring av partikler i et potensial. Sannsynligheten $P$ for å finne en partikkel i en gitt tilstand $(p,x)$, der $p$ og $x$ er henholdsvis partikkelens bevegelsesmengde og posisjon,

\begin{equation}
P \propto e^{-\beta E(p,x)},
\end{equation}

avhenger av partikkelens energi i tilstanden $(p,x)$ og $\beta = 1/(k_{B}T)$, der $k_{B}$ er Boltzmanns konstant og $T$ er partikkelens temperatur. Ved å anta at hver partikkel har samme bevegelsesmengde $p$ ved en gitt temperatur $T$, kan man anta at partikkelens kinetiske energi, $\frac{p^{2}}{2m}$, er konstant. Dermed vil sannsynligheten for å lokere en partikkel i en posisjon $x$ kun avhenge av $V(x)$ og $\beta$.

Ved å anta at partikkelen kun vil kunne bevege seg til venstre eller høyre under simuleringen, vil følgende sammenheng gjelde:

\begin{equation}
P^{+} + P^{-} = 1.
\end{equation}

Her er $P^{+}$ er sannsynligheten for å gå til høyre, og $P^{-}$ er sannsynligheten for å gå til venstre. 

Ut ifra denne sammenhengen finner man at sannsynligheten for at partikkelen beveger seg til venstre er 

\begin{equation}
P^{-} = \frac{e^{-\beta[V(x-h) - V(x+h)]}}{1+e^{-\beta[V(x-h) - V(x+h)]}}.
\end{equation}

 

Simuleringen av virrevandringen i et potensial utføres i funksjonen randomwalk_potensial og ligner på fremgangsmåten i funksjonen randow_walk i oppgave 3. Det slippes $n=1000$ partikler i samme posisjon $x = 0$, over $100$ tidssteg. Partiklene forflytter seg basert på et tilfeldig tall som sist i tillegg til formelen for sannsynlighet $P^{-}$ basert på potensialet. 
Denne funksjonen tar inn antall partikler, $n$ antall tidssteg, *timestep*, potensialet $v$, og parameteren $k$. Funksjonen returnerer en vektor av $n$ elementer, *pos*,  som inneholder posisjonen til hver partikkel etter simuleringen, i tillegg til forventningsverdi og standardavvik, *mean* og *std*, til posisjonen til en partikkel.

```python
def randomwalk_potential(n, timestep, v, k): #simulerer virrevandring av partikler i potensial
    pos = np.zeros(n) #vektor med posisjonene til n partikler

    for j in range(0,n - 1,1):
        
        for i in range(timestep):
            rr = np.random.uniform() #genererer tilfedlig tall
            pmin = 1 - 1/(1 + np.exp(-beta*(v(pos[j]-h,k)-v(pos[j]+h,k)))) 
            #sanns for at partikkelen forflytter seg enten til høyre eller til venstre
            
            if (rr >= pmin): #hvis det tilfeldige tallet er større enn sanns, forflytter partikkelen seg til høyre
                pos[j] += h
            else: #hvis det tilfeldige tallet er mindre enn sanns, forflytter partikkelen seg til venstre
                pos[j] -= h
    
    mean, std = norm.fit(pos) #beregner forventningsverdi og standardavvik til posisjonen til en partikkel
    
    return pos, mean, std                
                
```

Potensialet, som avhenger av posisjonen til partikkelen, er definert i funksjonen $v1$.
```python
#funksjon som returnerer potensialet som funksjon av posisjon. Tar inn posisjon, x, og verdi for k, k. 
def v1(x, k):
    return k*x
```

```python
'''
bruker funksjonen randomwalk_potential 1000 partiker, 
100 tidssteg, potensialet V(x)=kx og de tre ulike verdiene for beta*k, som nevnt over 
'''

pos2, mean2, std2 = randomwalk_potential(n5,timestep5, v1,k1) #med beta*k = 0.1
pos3, mean3, std3 = randomwalk_potential(n5,timestep5,v1,k2) #med beta*k = 1
pos4, mean4, std4 = randomwalk_potential(n5,timestep5, v1,k3) #med beta*k = 10
```

# 7.1

####  Antar $V_{Na} = V_K = V_0$

Det første som gjøres er å definerer flere konstanter gitt i oppgaveteksten.
Hvilepotensialet, $V_{rest}$, er potensialet i cellen når cellen hviler, og beregnes basert på startkonsentrasjonene på begge sider av membranen.
Kapasistansen, $C_c$, er også oppgit for gitte initialbetingelser,

\begin{equation}
\label{kapasistans}
C_c = \frac{[Q(t)]_{rest}^+}{V_{rest}}
\end{equation}

der $Q(t)_{rest}^+$ er konsentrasjonsforskjellen mellom innside og utside av cellen ved start.

De tre potensialene, her kalt $V1$, $V2$ og $V3$ er det tidsuavhengige membranpotensialet som i denne oppgaven antas likt for natrium-og kaliumpartiklene. Dette potensialet virker kun i membranområdet $x \in \, [-h,h]$ og omtales videre som $V0$. Det får tre ulike verdier, 

\begin{equation}
\beta V_0 = 0
\end{equation}


\begin{equation}
\beta V_0 = 1
\end{equation}

\begin{equation}
\beta V_0 = 3.
\end{equation}
 

Simuleringen starter med en posisjonsvektor hver til henholdsvis natriumpartiklene og kaliumpartiklene. Hver element i posisjonsvektoren inneholder som i tidligere oppgaver posisjonen til én partikkel. Indeksen strekker seg fra $\in \, [0,n]$, og angir partikkelnummer slik at vi kan bestemme antall partikler som skal slippes på utsiden eller innsiden av cellen ved start.

```python
#POTENSIALER
v_rest = -0.07 #gitt i volt 

#Membranpotensial: parametere som justeres
V1 = 0 #gitt i elektronvolt (eV)
V2 = 1/beta
V3 = 3/beta

#Kapasitans Cc (konstant basert på startkonsentrasjon)
cap = 0.07  # mol per liter*volt (M/V)
```

Det finnes to ulike potensialer i denne simuleringen. Membranpotensialet, $V_0$ som kun virker i membranenområdet som definert over, og et tidsavhengig, elektrisk potensial, $V(t)$.

Dette potensialet beregnes basert på konsentrasjonsforskjellen på hver side av membranen, og virker på både natrium- og kaliumpartiklene. Potensialet, videre omtalt som $V_{time}$, er lineært med hensyn på $x$ inni membranen

\begin{equation}
V_{time} =  \frac{[Q(t)]^+}{C_c}.
\label{V_time}
\end{equation}

$[Q(t)]^+$ angir konsentrasjonsforskjellen mellom innsiden og utsiden av cellen, og $C_c$ er kapasistansen. Vi antar som gitt i oppgavebeskrivelsen at antall partikler på utsiden av cellen holdes konstant under utregning av konsentrasjonsforskjellen mellom innsiden og utsiden. 


Funksjonen $Vt$ tar inn posisjonen til en partikkel $x$, det elektriske potensialet $V_{time}$, lengden på området $L$, skrittlengden $h$ og det tidsuavhengige potensialet $V0$. Den returnerer totalt potensial gitt partikkelens posisjon $x$. Vi har her valgt å sette $V_{time} = 0$ på utsiden av cellen $x_{outside} \in \, [h,\frac{L}{2}]$.

```python
#POTENSIALFUNKSJON

#Tidsavhengig, elektrisk potensial V_time som også er posisjonsavhengig
def Vt(x, V_time, L, h, V0):
    
    if (-h <= x and x <= h):
        return V0 + (-1/2*x+1/2)*V_time*e #potensialet i midten av membranen er V(t)/2
    if x < -h:
        return V_time*e
    if x > h:
        return 0
```

Det er også nødvendig å kunne telle antall partikler på innsiden av cellen for å beregne det elektriske potensialet $V_{time}$ ved hvert tidssteg senere.
Funksjonen count_part tar inn $L$, $h$ og en  posisjonsvektor, $pos$. Denne vektoren har som tidligere forklart en lengde lik antall partikler og inneholder verdien til posisjonen til hver partikkel.
Funksjonen teller så opp og returnerer antall partikler som befinner seg på innsiden av cellen. 

```python
def count_part(pos,L,h):
    n = 0
    for i in pos: #itererer gjennom hver partikkel i vektoren pos
        if i < (-h): #alle partikler på innsiden (-h)
            n += 1
    return n
```

Funksjonen under kalt *random_walk7*, er basert på liknende logikk som i tidligere oppgaver (3 og 5).
Inputparametere er posisjonsvektoren pos, potensialfunksjonen $V_t$ som også tar inn parameteren $V_{time}$ og $V_0$, som er forklart grundigere ved funksjonsdeklarasjonen. I tillegg tar den inn lengden av området $L$ og, skrittlengden $h$. 

Funksjonen itererer seg gjennom hver partikkel i vektoren $pos$, og regner ut sannsynligheten for hvilken retning partikkelen vil bevege seg i før den flytter partikkelen et skritt i gitt retning. Denne sannsynligheten, $P^{-}$, er nå avhengig av de to potensialene gitt ovenfor.
Funksjonen tar også hensyn til at partiklene ikke skal kunne bevege seg utenfor vårt koordinatsystem $x \in \, [-\frac{L}{2},\frac{L}{2}]$. Den returnerer så de oppdaterte posisjonene til hver partikkelen i *pos*-vektoren etter å ha løkket seg gjennom alle partiklene.

```python
def random_walk7(pos, V0, Vt, V_time, L, h):
    
    for i in range(len(pos)): #itererer gjennom hver partikkel
        r = np.random.uniform(0,1) #tilfeldig tall fra 0 til 1
        
        #Regner ut sannsynlighet for hvor partikkelen skal flyttes gitt av formel for P- fra oppg. 5
        exponent_left = Vt((pos[i]-h), V_time, L, h, V0) 
        exponent_right = Vt((pos[i]+h), V_time, L, h, V0)
        exponent = (-beta)*(exponent_left - exponent_right) 
        
        
        #pmin er sannsynlighet for å gå til venstre
        pMin = np.exp(exponent) / ( 1 +  np.exp(exponent) ) #sannsynligheten
        
        
        #Bestemmer partikkelens forflytningsretning
        #først: grensebetingelser ved endene
     
            
        if r <= pMin:  # skal bevege seg til venstre
            if pos[i] - 1 < -(L/2): 
                pos[i] += 1
            else: 
                pos[i] -= 1             
        if r > pMin:
            if pos[i] + 1 > (L/2): # øvre grense, kan ikke passere
                pos[i] -= 1            
            else:
                pos[i] += 1             
                        
        
    return pos
```

Funksjonen *potential* skal regne ut og returnere den elektriske potensialforskjellen mellom innsiden og utsiden av cellen, $\Delta V(t)$.

For å gjøre dette tar den inn en rekke inputparametere. $V_t$ er potensialfunksjonen som forklart i random_walk. Funksjonen tar også inn potensialene $V_{Na}$ og $V_{K}$ som virker på hver av partiklene i membranen. De antas like gjennom oppgave 7.1. Videre tar den inn antall tidssteg, her kalt timestep, $L$, og steglengde $h$.

Funksjonen returnerer en vektor med beregnede verdier for potensialforskjellen $\Delta V(t)$ for hvert tidssteg.


Det første som gjøres i funksjonen er å lage én posisjonsvektor hver til henholdsvis natriumpartiklene og kaliumpartiklene. I resten av prosjektet slippes $50$ natriumioner og $1400$ kaliumioner på innsiden av cellen $ (x = -L/4)$, og $50$ natriumioner og $1450$ kaliumioner på utsiden av cellen $(x=L/4)$, ved $t=0$. Hvert element i posisjonsvektoren inneholder som i tidligere oppgaver posisjonen til én partikkel. Indeksen strekker seg fra $\in \, [0,n]$. Den angir partikkelnummer slik at vi kan bestemme antall partikler som skal slippes på utsiden eller innsiden av cellen ved start.
Videre beregnes den elektriske potensialforskjellen mellom membranveggene, her omtalt som $V_{time}$. 

```python
def potential(Vt, V0_Na, V0_K, timestep, L, h):
    
    pos_Na = np.zeros(n_Na) #Liste for Na-partikler
    pos_Na[(n_Na_i):(n_Na)] = L/4 #slipper partiklene utenfor cellen
    pos_Na[0: n_Na_i] = -L/4 #slipper partikler inni cellen
    

    pos_K = np.zeros(n_K) #Liste for Na-partikler
    pos_K[(n_K_i):(n_K)] = L/4 #slipper partiklene utenfor cellen
    pos_K[0: n_K_i] = -L/4 #slipper partikler inni cellen
    
    deltaV_list = [] #lager en tom liste der deltaV fra hvert tidssteg skal inn
    
    #itererer gjennom antall tidssteg
    for i in range(timestep):
        n_inside = count_part(pos_Na, L, h) + count_part(pos_K, L, h) #teller opp antall partikler på innsiden
    
        #regner ut potensialet V_time i området
        Q = n_inside - n_out
        V_time = (Q*1e-4)/cap #konsentrasjonsforskjellen
        
        deltaV_list.append(V_time)
    
        #utfører random walk og returnerer oppdatert posisjonsvektor for Na og K
        pos_Na = random_walk7(pos_Na, V0_Na, Vt, V_time, L, h)
        pos_K = random_walk7(pos_K, V0_K, Vt, V_time, L, h)
        
    
    return np.array(deltaV_list)*1000
```

#### Denne oppgaven omhandler simulering av aksjonspotensialet

I denne oppgaven bruker vi samme tankegang og logikk i kodingen som i oppgave 7, men denne gangen legger vi til spennings-kontrollerte membranpotensialer. $V_{Na}(x,t)$ og $V_{K}(x,t)$ er ikke lenger kun posisjonsavhengige potensialer, men nå også tidsavhengige. # 8.1
#### Uten $Na^{+}/K^{+}$-pumpen (SSP)

De lukkede portene har potensial $\beta V_{Na}^{0} = \beta V_{K}^{0} = 50$, mens de åpnede portene har potensial $\beta V_{open} = 1$. Simuleringen baserer seg på de samme startkonsentrasjonene som oppgave 7.

Når membranspenningen $V(t) \leq - 70 mV$, settes Na-portens potensial til $V_{open}$ i membranen, slik at Na-ioner kan strømme inn i cellen. Na-porten stenges når $V(t)$ når $30 mV$, ved at $V_{Na}$ settes til $\beta V_{Na}^{0}$ i membranen. Samtidig åpnes K-porten ved at $\beta V_{K}$ settes til $V_{open}$ i membranen. K-ioner strømmer over til utsiden av cellen til membranpotensialet igjen senkes til $-70 mV$. Da stenges K-porten og Na-porten åpnes igjen, og denne prosessen vil fortsette i en syklus. 

Potensialet er definert i to ulike funksjoner, *potentialNa* og *potentialK*, for henholdsvis Na-ioner og K-ioner. Disse funksjonen returnerer potensialet til en partikkel i en gitt posisjon, til et gitt tidspunkt. Funksjonene tar hensyn til om portene til K og Na er åpne eller ikke. 

Simuleringen kjøres over 1000 tidssteg, med Δt = 1, og virrevandringen gjøres i *random_walk8*. Den eneste forskjellen på denne funksjonen og *randomwalk_potential*, er at vi nå har ulike potensialfunksjoner for de to ulike partikkeltypene. 

```python
# Bruker samme konsentrasjoner som i oppgave 7
timestep_81 = 1000

V_K = 50/beta
V_Na = 50/beta

V_Na_open = 1/beta
V_K_open = 1/beta

V_Na_closed = 50/beta
V_K_closed = 50/beta

V_start = ((-50)*1.0e-04)/cap
```

```python
#Regner ut potensialet til Na i barrieren


def potentialNa(x, V_time):
    global V_Na
    global V_K
    if (x <= h and x >= -h): #i barriereområdet
        if V_time >= 0.03: #
            V_K = V_K_open #åpner K-porten og lukker Na-porten 
            V_Na = V_Na_closed
            return (-1/2*x+1/2)*V_time*e + V_Na 
        if V_time <= V_start: #åpner Na porten, og lukker K-porten
            V_K = V_K_closed
            V_Na = V_Na_open
            return  (-1/2*x+1/2)*V_time*e + V_Na 
        else:
            return (-1/2*x+1/2)*V_time*e + V_Na 
    if x < -h:
        return V_time*e
    if x > h:
        return 0
    
    
#Regner ut potensialet til K i barrieren
def potentialK(x, V_time):
    global V_Na
    global V_K
    if x <= h and x >= -h: #i barriereområdet
        return V_K + (-1/2*x+1/2)*V_time*e
    if x < -h:
        return V_time*e
    if x > h:
        return 0
    
```

```python
def random_walk8(pos, Vt, V_time):
    
    for i in range(len(pos)): #itererer gjennom hver partikkel
        r = np.random.uniform(0,1) #tilfeldig tall fra 0 til 1
        
        #Regner ut sannsynlighet for hvor partikkelen skal flyttes gitt av formel for P- fra oppg. 5
        exponent_left = Vt(pos[i]-h, V_time) 
        exponent_right = Vt(pos[i]+h, V_time)
        exponent = (-beta)*(exponent_left - exponent_right) 
        
        
        #pmin er sannsynlighet for å gå til venstre
        pMin = np.exp(exponent) / ( 1 +  np.exp(exponent) ) #sannsynligheten
        
        
        
        #Bestemmer partikkelens forflytningsretning
        #først: grensebetingelser ved endene
     
        if r <= pMin:  # skal bevege seg til venstre
            if pos[i] - 1 < -(L/2): 
                pos[i] += 1
            else: 
                pos[i] -= 1             
        if r > pMin:
            if pos[i] + 1 > (L/2): # øvre grense, kan ikke passere
                pos[i] -= 1            
            else:
                pos[i] += 1             
                        
        
    return pos 
```

```python
#Beregner spenningsforskjellen i cellemembranen
def potential_8(Vt_Na, Vt_K, timestep, L, h):
    
    pos_Na = np.zeros(n_Na) #Liste for Na-partikler
    pos_Na[(n_Na_i):(n_Na)] = L/4 #slipper partiklene utenfor cellen
    pos_Na[0: n_Na_i] = -L/4 #slipper partikler inni cellen
    

    pos_K = np.zeros(n_K) #Liste for Na-partikler
    pos_K[(n_K_i):(n_K)] = L/4 #slipper partiklene utenfor cellen
    pos_K[0: n_K_i] = -L/4 #slipper partikler inni cellen
    
    deltaV_list = [] #lager en tom liste der deltaV fra hvert tidssteg skal inn
    
    #itererer gjennom antall tidssteg
    for i in range(timestep):
        n_inside = count_part(pos_Na, L, h) + count_part(pos_K, L, h) #teller opp antall partikler på innsiden
    
        #regner ut potensialet V_time i området
        Q = n_inside - n_out
        V_time = (Q*1e-4)/cap #konsentrasjonsforskjellen
        
        
        deltaV_list.append(V_time)
    
        #utfører random walk og returnerer oppdatert posisjonsvektor for Na og K
        pos_Na = random_walk8(pos_Na, Vt_Na, V_time)
        pos_K = random_walk8(pos_K, Vt_K, V_time)
        
    
    return np.array(deltaV_list)*1000
```

Tanken bak simuleringen i denne oppgaven er lik som i oppgave 8.1. Det eneste som er forandret, er at vi har lagt til $Na^{+}/K^{+}$-pumpen (SSP). Denne pumpen sender over 2 Na-ioner fra utsiden til innsiden av cellen, og 3 K-ioner fra innsiden til utsiden av cellen for hvert tiende tidssteg. Dersom det ikke er nok Na-ioner eller K-ioner å ta av, vil ikke SSP kjøre. Dette er implementert i funksjonen *potential_82*, som beregner potensialet på tvers av cellemembranen som funksjon av tiden. 

Simuleringen kjøres over $5000$ tidssteg.

```python
def potential_82(Vt_Na, Vt_K, timestep, L, h):
    
    pos_Na = np.zeros(n_Na) #Liste for Na-partikler
    pos_Na[(n_Na_i):(n_Na)] = L/4 #slipper partiklene utenfor cellen
    pos_Na[0: n_Na_i] = -L/4 #slipper partikler inni cellen
    

    pos_K = np.zeros(n_K) #Liste for Na-partikler
    pos_K[(n_K_i):(n_K)] = L/4 #slipper partiklene utenfor cellen
    pos_K[0: n_K_i] = -L/4 #slipper partikler inni cellen
    
    deltaV_list = [] #lager en tom liste der deltaV fra hvert tidssteg skal inn
    
    count = 0
    
    #itererer gjennom antall tidssteg
    for i in range(timestep):
        nNa_inside = count_part(pos_Na, L, h)
        nK_outside = 1450 - count_part(pos_K, L, h)
        
        if nNa_inside > 3 and nK_outside > 2:
            count += 1
        else:
            count = 0
            
        if (count == 10):
            if nNa_inside > 3 and nK_outside > 2: #flytter Na
                n = 0
                k = 0
                while k < 3:
                    if pos_Na[k+n] < -0.5:
                        pos_Na[k+n] = -pos_Na[k+n]
                        k += 1
                    else:
                        n += 1
                n=0
                k=0
                
                while k < 2: #flytter K
                    if pos_K[k+n] > 0.5:
                        pos_K[k+n] = -pos_K[k+n]
                        k += 1
                    else:
                        n += 1
                count = 0
            else:
                count = 0
    
        
        
        n_inside = count_part(pos_Na, L, h) + count_part(pos_K, L, h) #teller opp antall partikler på innsiden
    
        #regner ut potensialet V_time i området
        Q = n_inside - n_out
        V_time = (Q*1e-4)/cap #konsentrasjonsforskjellen
       # print("V_time:", V_time)
        
        deltaV_list.append(V_time)
    
        #utfører random walk og returnerer oppdatert posisjonsvektor for Na og K
        pos_Na = random_walk8(pos_Na, Vt_Na, V_time)
        pos_K = random_walk8(pos_K, Vt_K, V_time)
        
    
    return np.array(deltaV_list)*1000
```
